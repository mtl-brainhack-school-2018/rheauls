Programme en Python pour la generation de donnees selon une distribution normale ou uniforme

Module 1 - Generation des donnees

```python
# Module pour generer les variables t,u,v,w,x,y et z
# cognition = f(t,u,v,w,x,y,z) = 2.1*t/140 + 3.6*u/70 + 2.8*v/7 + 5*w/80 + 1.4*x/5 - 2.5*y/12 + 4*z/40
#   t = tasys=systolic blood pressure (moyenne=140, deviation standard=20)
#   u = tadia=diastolic blood pressure (moyenne=70, deviation standard=10)
#   v = hba1c=glycosylated Hb (moyenne=7, deviation standard=2)
#   w = age=age (moyenne=80, deviation standard=10)
#   x = ldl=LDL (moyenne=5, deviation standard=2)
#   y = edu=years of education (moyenne=12, deviation standard=2)
#   z = vs=VS (moyenne=40, deviation standard=20)


import numpy as np
import math


#
# entrez le nombre de participants (nb)
# deux versions: 3 premieres lignes pour l'entrer au clavier OU l'entrer dans le code directement
#v1-1 print('Entrez le nombre de participants: ')
#v1-2 input(nb_part)
#v1-3 nb = int(nb_part)
#OU

nb = 1000

a = np.zeros((nb, 1))
b = np.zeros((nb, 1))
c = np.zeros((nb, 1))
d = np.zeros((nb, 1))
e = np.zeros((nb, 1))
f = np.zeros((nb, 1))
g = np.zeros((nb, 1))


#
# TA systolique (tasys)

compteur = 0
while compteur < nb:
    a[compteur] = math.floor(np.random.normal(140.0, 20.0, 1))
    compteur+=1

tasys = a
#print(tasys)

print('moyenne TA systolique: ', np.mean(tasys))
print('deviation standard TA systolique: ', np.std(tasys))
print('')


#
# TA diastolique (tadia)
compteur = 0
while compteur < nb:
    b[compteur] = math.floor(np.random.normal(70, 10.0, 1))
    compteur+=1

tadia = b
#print(tadia)

print('moyenne TA diastolique: ', np.mean(tadia))
print('deviation standard TA diastolique: ', np.std(tadia))
print('')


#
# HbA1C (hba1c)
compteur = 0
while compteur < nb:
    c[compteur] = np.random.normal(7, 2, 1)
    compteur+=1

hba1c = c
#print(hba1c)

print('moyenne hemoglobine glyquee HbA1C: ', np.mean(hba1c))
print('deviation standard hemoglobine glyquee HbA1C: ', np.std(hba1c))
print('')


#
# Age (age)
compteur = 0
while compteur < nb:
    d[compteur] = math.floor(np.random.normal(80, 10.0, 1))
    compteur+=1

age = d
#print(age)

print('moyenne age: ', np.mean(age))
print('deviation standard age: ', np.mean(age))
print('')


#
# Cholesterol LDL (ldl)
compteur = 0
while compteur < nb:
    e[compteur] = np.random.normal(5, 2, 1)
    compteur+=1

ldl = e
#print(ldl)

print('moyenne cholesterol LDL:', np.mean(ldl))
print('deviation standard cholesterol LDL: ', np.std(ldl))
print('')


#
# Annees d'education (edu)
compteur = 0
while compteur < nb:
    f[compteur] = math.floor(np.random.normal(12, 2, 1))
    compteur+=1

edu = f
#print(edu)

print('moyenne education: ', np.mean(edu))
print('deviation standard education: ', np.std(edu))
print('')


#
# Vitesse de sedimentation/inflammation (vs)
compteur = 0
while compteur < nb:
    g[compteur] = math.floor(np.random.normal(40, 20, 1))
    if g[compteur]<0:
        g[compteur] = -g[compteur]
    else:
        g[compteur] = g[compteur]
    compteur+=1
    
vs = g
#print(vs)

print('moyenne vitesse de sedimentation: ', np.mean(vs))
print('deviation standard vitesse de sedimentation: ', np.std(vs))
print('')

print(tasys)
print(tadia)
print(hba1c)
print(age)
print(ldl)
print(edu)
print(vs)
```

# Calcul de la cognition selon la formule dans l'entete

```python
#
# cognition (cog)
h = np.zeros((nb, 1))

compteur = 0
while compteur < nb:
    h[compteur] = - 2.1*(tasys[compteur]/110) - 3.6*(tadia[compteur]/60) - 2.8*(hba1c[compteur]/7) - 5*(age[compteur]/65) - 1.4*(ldl[compteur]/4) + 2.5*(edu[compteur]/10) - 4*(vs[compteur]/30)
    compteur+=1

cog = h
#print(cog)

print('moyenne cognition: ', np.mean(cog))
print('deviation standard cognition: ', np.std(cog))
print('')

type(cog)

compteur = 0
while compteur < nb:
    cog[compteur] = np.round(cog[compteur], 2)
    compteur+=1

print(cog)
```

# Creation du fichier contenant les donnees en format csv

```python
import pandas as pd


#
# Inclusion des donnees dans un tableau nomme data_table
# Creation du fichier bh_data_table.csv


#
# reinitialise le tableau met les donnees aleatoires calculees dans les cellules precedentes

data_list = list()

compteur_participant = 0
while compteur_participant < nb:
    participant_data = list()
    participant_data.append(tasys[compteur_participant])
    participant_data.append(tadia[compteur_participant])
    participant_data.append(hba1c[compteur_participant])
    participant_data.append(age[compteur_participant])
    participant_data.append(ldl[compteur_participant])
    participant_data.append(edu[compteur_participant])
    participant_data.append(vs[compteur_participant])
    participant_data.append(cog[compteur_participant])
    data_list.append(participant_data)
    compteur_participant+=1

          
# 
# Creation du fichier cvs a partir de Pandas
data_list_entete = ['tasys','tadia','hba1c','age','ldl','edu','vs','cog']
bh_data_table = pd.DataFrame(data=data_list)
print(bh_data_table)
bh_data_table.to_csv('bh_data_table.csv', header=data_list_entete)

```
